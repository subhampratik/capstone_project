#### 11.Using a user defined function convert the ‘emp_len’ column into categorical column as follows -

1. If emp_len is less than equals to 1 then recode as ‘fresher’.
2. If emp_len is greater than 1 and less than 3 then recode as ‘junior’.
3. If emp_len is greater than 3 and less than 7 then recode as ‘senior’
4. If emp_len is greater than 7 then recode as ‘expert’.

```python
loan['emp_length'] = loan['emp_length'].str.extractall('(\d+)').unstack().astype(float)
loan
```

```python
loan.emp_length.dtypes
```

```python
def user_def_func(n):
    if n<=1:
        return 'fresher'
    if n>1 and n<=3:
        return 'junior'
    if n>=3 and n<7:
        return 'senior'
    if n>=7:
        return 'expert'
loan["emp_length"] = loan["emp_length"].apply(user_def_func)
```
